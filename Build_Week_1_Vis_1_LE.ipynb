```
# Import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```
dr_le = pd.read_csv('https://raw.githubusercontent.com/KristineYW/Unit-1-Build/master/NCHS_-_Death_rates_and_life_expectancy_at_birth.csv')

dr_le.head()
```

```
dr_le_all = dr_le.drop(['Race', 'Age-adjusted Death Rate'], axis=1)

dr_le_all.head(10)
```

```
dr_le_all = dr_le.drop(['Race', 'Age-adjusted Death Rate'], axis=1)

dr_le_all.head(10)
```

```
# Create new dataframe with just total population information

dr_le_pop = dr_le.loc[dr_le['Race'] == 'All Races']
```

```
# Delete gendered rows

dr_le_pop = dr_le_pop.loc[dr_le_pop['Sex'] == 'Both Sexes']

dr_le_pop.head()
```

```
# Delete Race, Sex, and Age-adjusted Death Rate columns now that they contain only one value

dr_le_pop = dr_le_pop.drop(['Sex','Race','Age-adjusted Death Rate'], axis=1)

dr_le_pop.head()
```

```
# Check the dataset for NaNs

dr_le_pop.describe()
```

```
# Create Visualization



fig, ax = plt.subplots(figsize=(18, 10))

ax.set_title("Average Life Expectancy from 1900-2017", color='black', weight='bold', fontsize=24, x=.475, y=1.05)

ax.set_yticklabels([35,40,45,50,55,60,65,70,75,80], color='#616161', fontsize=15)
ax.set_xticklabels([1880,1900,1920,1940,1960,1980,2000,2020], color='#616161', fontsize=15)

ax.grid()
a = ax.get_ygridlines()

fig.set_facecolor('#f0f0f0')
ax.set_facecolor('#f0f0f0')

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

ax.plot(dr_le_pop['Year'],dr_le_pop['Average Life Expectancy (Years)'], color='green',linewidth=3);
```

```
# Create race-specific dataset

dr_le_by_race = dr_le.loc[dr_le['Sex'] == 'Both Sexes']

dr_le_by_race.head(10)
```

```
# Drop Sex & Age-adjusted Death Rate column

dr_le_by_race = dr_le_by_race.drop(['Sex','Age-adjusted Death Rate'], axis=1)

dr_le_by_race.head(10)
```

```
# Filter out data for Race: white

dr_le_white = dr_le_by_race[dr_le['Race'] == 'White']

dr_le_white.head(10)
```

```
# Filter out data for Race: black

dr_le_black = dr_le_by_race[dr_le['Race'] == 'Black']

dr_le_black.head(10)
```

```
# Create gendered dataframe


dr_le_by_sex = dr_le.loc[dr_le['Race'] == 'All Races']

dr_le_by_sex.tail(10)
```

```

dr_le_by_sex = dr_le_by_sex.drop(['Race','Age-adjusted Death Rate'], axis=1)

dr_le_by_sex.tail(10)
```

```
dr_le_m = dr_le_by_sex[dr_le['Sex'] == 'Male']

dr_le_m.head(100)
```

```
dr_le_f = dr_le_by_sex[dr_le['Sex'] == 'Female']

dr_le_f.head(10)
```

```
# Create Visualization



fig, ax = plt.subplots(figsize=(18, 10))

ax.set_title("Average Life Expectancy (Years) from 1900-2017", color='black', weight='bold', fontsize=24, x=.475, y=1.05)
fig.text(x=.28,y=.17,s="*WWI", fontsize=15)

ax.set_yticklabels([20,30,40,50,60,70,80], color='#000000', fontsize=20)
ax.set_xticklabels([1880,1900,1920,1940,1960,1980,2000,2020], color='#000000', fontsize=20)


fig.set_facecolor('#FFFFFF')
ax.set_facecolor('#FFFFFF')

ax.set_xlabel('Year',fontsize=25,labelpad=30)
ax.set_ylabel('Life Expectancy',fontsize=25,labelpad=20)

l1 = ax.plot(dr_le_pop['Year'],dr_le_pop['Average Life Expectancy (Years)'], color='#000000',linewidth=5);
l2 = ax.plot(dr_le_white['Year'],dr_le_white['Average Life Expectancy (Years)'], color='#189ad3',linewidth=3);
l3 = ax.plot(dr_le_black['Year'],dr_le_black['Average Life Expectancy (Years)'], color='#107dac',linewidth=3);
l4 =ax.plot(dr_le_m['Year'],dr_le_m['Average Life Expectancy (Years)'], color='#A270A7',linewidth=3);
l5 = ax.plot(dr_le_f['Year'],dr_le_f['Average Life Expectancy (Years)'], color='#85478B',linewidth=3);


ax.legend((l1,l2,l3,l4,l5),labels=('All','White','Black','Male','Female'), 
          fontsize=22,bbox_to_anchor=(.95,.5),shadow=True,facecolor='#f1f1f1',borderpad=1)

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

ax.plot(dr_le_white['Year'],dr_le_white['Average Life Expectancy (Years)'], color='#189ad3',linewidth=3);
ax.plot(dr_le_black['Year'],dr_le_black['Average Life Expectancy (Years)'], color='#107dac',linewidth=3);
ax.plot(dr_le_m['Year'],dr_le_m['Average Life Expectancy (Years)'], color='#A270A7',linewidth=3);
ax.plot(dr_le_f['Year'],dr_le_f['Average Life Expectancy (Years)'], color='#85478B',linewidth=3);
ax.plot(dr_le_pop['Year'],dr_le_pop['Average Life Expectancy (Years)'], color='#000000',linewidth=5);
```
