```
# Import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```
ch_mort = pd.read_csv('https://raw.githubusercontent.com/KristineYW/Unit-1-Build/master/NCHS_-_Childhood_Mortality_Rates.csv')

ch_mort.head()
```

```
ch_mort['Death Rate'] = ch_mort['Death Rate'].str.replace(',','').astype(float)

ch_mort.head()
```

```
ch_mort_p = ch_mort.pivot(index='Age Group',columns='Year',values='Death Rate')

ch_mort_p.head()
```

```
fig, ax = plt.subplots(figsize=(18, 10))

ax.set_title("Child Mortality Rates", color='black', weight='bold', fontsize=24, x=.475, y=1.05)

#ax.set_yticklabels([0,200,400,2000], color='#000000', fontsize=15)
ax.set_xticklabels([1880,1900,1920,1940,1960,1980,2000,2020], color='#000000', fontsize=15)


fig.set_facecolor('#FFFFFF')
ax.set_facecolor('#FFFFFF')

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

ax.plot(ch_mort['Year'],ch_mort['Death Rate'], color='green',linewidth=3);
```

```
# Pull out each age group into separate dataframes

first = ch_mort.loc[ch_mort['Age Group'] == '1-4 Years']

first.tail()
```

```
second = ch_mort.loc[ch_mort['Age Group'] == '5-9 Years']

second.head()
```

```
third = ch_mort.loc[ch_mort['Age Group'] == '10-14 Years']

third.head()
```

```
fourth = ch_mort.loc[ch_mort['Age Group'] == '15-19 Years']


fourth.head()
```

```
fig, ax = plt.subplots(figsize=(18, 10))

ax.set_title("Child Mortality Rates", color='black', weight='bold', fontsize=35, x=.475, y=1.15)
fig.text(x=.25,y=.95,s="Number of child deaths per 100,000 by age group", fontsize=25)
fig.text(x=.28,y=.69,s="*WWI", fontsize=15)

ax.set_yticklabels([0,0,250,500,750,1000,1250,1500,1750,2000], color='#000000', fontsize=15)
ax.set_xticklabels([1880,1900,1920,1940,1960,1980,2000,2020], color='#000000', fontsize=15)


fig.set_facecolor('#FFFFFF')
ax.set_facecolor('#FFFFFF')

ax.set_xlabel('Year',fontsize=25,labelpad=30)
ax.set_ylabel('Deaths per 100,000',fontsize=25,labelpad=20)

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

l1 = ax.plot(first['Year'],first['Death Rate'], color='#0e1f4c',linewidth=4);
l2 = ax.plot(second['Year'],second['Death Rate'], color='#17396c',linewidth=4);
l3 = ax.plot(third['Year'],third['Death Rate'], color='#245b96',linewidth=4);
l4 = ax.plot(fourth['Year'],fourth['Death Rate'], color='#6597b0',linewidth=4);

ax.legend((l1,l2,l3,l4),labels=('1-4 Years','5-9 Years','10-14 Years','15-19 Years'), 
          fontsize=22,bbox_to_anchor=(0.9,0.9),shadow=True,facecolor='#f1f1f1',borderpad=1)

ax.plot(first['Year'],first['Death Rate'], color='#0e1f4c',linewidth=3);
ax.plot(second['Year'],second['Death Rate'], color='#17396c',linewidth=3);
ax.plot(third['Year'],third['Death Rate'], color='#245b96',linewidth=3);
ax.plot(fourth['Year'],fourth['Death Rate'], color='#6597b0',linewidth=3);

plt.plot();
```
